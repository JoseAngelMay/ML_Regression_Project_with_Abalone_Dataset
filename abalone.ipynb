```python
# from this, we note that the following should be square-rooted to reduce skew:
# "Height", ""Whole weight", "Shuckled weight", "Viscera weight", "Shell weight"
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer, make_column_transformer, make_column_selector
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_ratio, feature_names_out = ratio_name),
        StandardScaler())

def column_ratio_alt(X):
    return X[:, [0]] / (X[:, [1]] * X[:, [2]] * X[:, [3]])

def ratio_alt_name(function_transformer, feature_names_in):
    return ["ratio_alt"]

def ratio_alt_pipeline():
    return make_pipeline(
        KNNImputer(),
        FunctionTransformer(column_ratio_alt, feature_names_out = ratio_alt_name),
        StandardScaler())

square_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.square, feature_names_out = "one-to-one"),
    StandardScaler())

sqrt_pipeline = make_pipeline(
    KNNImputer(),
    FunctionTransformer(np.sqrt, feature_names_out = "one-to-one"),
    StandardScaler())

default_num_pipeline = make_pipeline(KNNImputer(),
                                     StandardScaler())

cat_pipeline = make_pipeline(SimpleImputer(strategy = "most_frequent"),
                             OneHotEncoder(handle_unknown = "ignore"))

preprocessing = ColumnTransformer([
    ("Shellfulness", ratio_pipeline(), ["Shell weight", "Whole weight"]),
    ("Meattiness", ratio_pipeline(), ["Shucked weight", "Whole weight"]),
    ("Gutsy", ratio_pipeline(), ["Viscera weight", "Whole weight"]),
    ("Density", ratio_pipeline(), ["Whole weight", "Diameter", "Length", "Height"]),
    ("square", square_pipeline, ["Length", "Diameter"]),
    ("sqrt", sqrt_pipeline, ["Height", "Whole weight", "Shucked weight", "Viscera weight", "Shell weight"]),
    ("cat", cat_pipeline, make_column_selector(dtype_include = object)),
],
                                  remainder = default_num_pipeline)
```

```python
abalone_prepared = preprocessing.fit_transform(abalone)
abalone_prepared.shape
```

```python
preprocessing.get_feature_names_out()
```

# Fine-Tune Hyperparameters

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
lin_reg.get_params().keys()
```

```python
param_distribs = {'columntransformer__sqrt__knnimputer__n_neighbors': randint(low = 1, high = 9),
                  'columntransformer__square__knnimputer__n_neighbors': randint(low = 1, high = 9),
                  'columntransformer__cat__simpleimputer__strategy': ["most_frequent"]}
rnd_search_lin = RandomizedSearchCV(
    lin_reg, param_distributions = param_distribs, n_iter = 10, cv = 5,
    scoring = "neg_root_mean_squared_error", random_state = 42)

rnd_search_lin.fit(abalone, abalone_labels)
```

```python
dec_tree.get_params().keys()
```

```python
param_distribs = {'decisiontreeregressor__max_depth': randint(low = 1, high = 15),
                  'decisiontreeregressor__max_leaf_nodes': randint(low = 2, high = 15),
                  'decisiontreeregressor__min_samples_leaf': randint(low = 2, high = 25)}
rnd_search_dt = RandomizedSearchCV(
    dec_tree, param_distributions = param_distribs, n_iter = 10, cv = 5,
    scoring = "neg_root_mean_squared_error", random_state = 42)

rnd_search_dt.fit(abalone, abalone_labels)
```

```python
svr.get_params().keys()
```

```python
param_distribs = {'svr__degree': randint(low = 1, high = 9)}

rnd_search_svr = RandomizedSearchCV(
    svr, param_distributions = param_distribs, n_iter = 10, cv = 5,
    scoring = "neg_root_mean_squared_error", random_state = 42)

rnd_search_svr.fit(abalone, abalone_labels)
```

```python
lin_final = rnd_search_lin.best_estimator_
lin_rmse = -cross_val_score(lin_final, abalone, abalone_labels, scoring = "neg_root_mean_squared_error", cv = 10)
pd.Series(lin_rmse).describe()
```

```python
dec_tree_final = rnd_search_dt.best_estimator_
dt_rmse = -cross_val_score(dec_tree_final, abalone, abalone_labels, scoring = "neg_root_mean_squared_error", cv = 10)
pd.Series(dt_rmse).describe()
```

```python
svr_final = rnd_search_svr.best_estimator_
svr_rmse = -cross_val_score(svr_final, abalone, abalone_labels, scoring = "neg_root_mean_squared_error", cv = 10)
pd.Series(svr_rmse).describe()
```

```python
from sklearn.ensemble import VotingRegressor

vot_reg = VotingRegressor(
    estimators = [
        ('lin_reg', rnd_search_lin.best_estimator_),
        ('dec_tree', rnd_search_dt.best_estimator_),
        ('svr', rnd_search_svr.best_estimator_)
    ]
)
vot_rmse = -cross_val_score(vot_reg, abalone, abalone_labels, scoring = "neg_root_mean_squared_error", cv = 10)
pd.Series(vot_rmse).describe()
```
